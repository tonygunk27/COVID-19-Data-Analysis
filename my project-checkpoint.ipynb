```python
data=pd.read_csv("covid19_Confirmed_dataset.csv")
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
data.head(10)
```

```python
cagg=data.groupby("Country/Region").sum()
```

```python
cagg.head()
```

```python
cagg.shape
```

```python
cagg.head(10)
```

```python
cagg.loc["China"].plot()
plt.legend()
```

```python
cagg.loc["China"][:3].plot()
plt.legend()
```

```python
cagg.loc["China"].diff().plot()
plt.legend()
```

```python
cagg.loc["China"].diff().max()
```

```python
countries=list(cagg.index)
max=[]
for c in countries:
    max.append(cagg.loc[c].diff().max())
```

```python
max
```

```python
cagg["max"]=max
```

```python
cagg.head(10)
```

```python
data=pd.DataFrame(cagg["max"])
```

```python
data.head(10)
```

```python
happ=pd.read_csv("worldwide_happiness_report.csv")
```

```python
happ.head(10)
```

```python
useless_col=["Perceptions of corruption", "Generosity", "Freedom to make life choices", "Social support"]
happ.drop(useless_col, axis=1, inplace=True)
```

```python
happ.head(10)
```

```python
happ.set_index("Country or region", inplace=True)
```

```python
happ.head(10)
```

```python
data.shape
```

```python
happ.shape
```

```python
df=data.join(happ, how="inner")
```

```python
df.head(20)
```
